```python
import random as rm
import numpy as np
import matplotlib.pyplot as plt
import pylab as pl
import math

%matplotlib inline
##############################
def plotProb(num):
    data=[rm.uniform(-np.pi/2,np.pi/2.) for i in xrange(num)]
    
    beta=1.0
      
    h=y=0.0
    prod=1.0
    
    alpha=1.
    h=np.linspace(-6,6,num)
    s=l=np.zeros(num)
    
    xk=np.tan(data)*beta+alpha
    xk[:3]=[4.9,0.1,-2.]
    
    for n in xrange(len(data)):
        for x in xrange(n):
            s+= np.log((beta**2)+((xk[x]-h)**2))
        #l+= np.log((beta**2)+((xk[n]-h)**2))
        l=s
        s=0.
        l*=-1.
        l-=np.max(l)
        l=np.exp(l)
        if(n>0):
            mean=np.mean(xk[:n])
        
        if(n==1 or n==2 or n==3 or n == 8):# or n==256):
            plt.title('n= '+str(n)+' ')
            plt.plot(h,l,label='n=')
            plt.axvline(x=mean,linestyle='--')
            plt.plot(xk[:n],np.ones(n)*1.001,'o')
            plt.ylabel(r'prob($\alpha$|{$x_k$},$\beta$,I)')
            plt.xlabel(r'Lighthouse position $\alpha$ (km)')
            plt.show()
        
    #print np.max(xk)
    plt.title('Final Plot')    
    plt.plot(h,np.exp(l))
    plt.axvline(x=mean,linestyle='--')
    plt.ylabel(r'prob($\alpha$|{$x_k$},$\beta$,I)')
    plt.xlabel(r'Lighthouse position $\alpha$ (km)')
    plt.show()
    
    print 'position of max: ',h[np.argmax(np.exp(l))]
##############################
#Running the code    
plotProb(512)
```
